```python
sum(data.query("alive == 'no'" )['fare'])
```

**Survivals based on class**

```python
fcSurvived = len(data.query("pclass == 1 and alive =='yes'"))
scSurvived = len(data.query("pclass == 2 and alive =='yes'"))
tcSurvived = len(data.query("pclass == 3 and alive =='yes'"))
totalSurvived = len(data.query("alive == 'yes'"))
print("First Class Survived :", fcSurvived, "Percentage = ",fcSurvived/(totalSurvived)*100)
print("Second Class Survived :", scSurvived, "Percentage = ",scSurvived/(totalSurvived)*100)
print("Third Class Survived :", tcSurvived, "Percentage = ",tcSurvived/(totalSurvived)*100)
```

**is there any relation ship between fare and survival**



```python
print("Southampton Survived : ",len(data.query("embark_town == 'Southampton' and alive=='yes'")))
print("Queenstown Survived : ",len(data.query("embark_town == 'Queenstown' and alive=='yes'")))
print("Cherbourg Survived : ",len(data.query("embark_town == 'Cherbourg' and alive=='yes'")))
```

## MatplotLib - Data Visualisation

```python
x = [1,5,10,3,2]
y = [1,10,1,6,5]
plt.plot(x,y) # a continous relation between data points
plt.scatter(x,y)  # data points plotted
plt.bar(x,y) #
# plt.pie(x,y)
# plt.hist(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("My first Graph")
```
